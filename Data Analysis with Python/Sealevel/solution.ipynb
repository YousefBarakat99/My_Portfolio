```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
df = pd.read_csv('epa-sea-level.csv')
df.loc[df['Year'] == 2000]
```

```python
fig, axes = plt.subplots(figsize=(14,6))
axes.set_xlabel('Year')
axes.set_ylabel('Sea Level (inches)')
axes.set_title('Rise in Sea Level')
axes.scatter(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])
```

```python
linregress(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])
```

```python
plt.figure(figsize=(14,6))
ax = plt.axes()
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
ax.scatter(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])
y2 = (0.06304458401213482 * 2050) - 119.06594196773983
ax.plot([1880, 2050], [0, y2], color='red')
plt.show()
```

```python
recent = df.copy()
recent = recent.loc[recent['Year']>=2000]
recent = recent.reset_index()
recent = recent.drop(columns='index')
linregress(x=recent['Year'], y=recent['CSIRO Adjusted Sea Level'])
```

```python
plt.figure(figsize=(14,6))
ax = plt.axes()
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Rise in Sea Level')
ax.scatter(x=df['Year'], y=df['CSIRO Adjusted Sea Level'])
y2 = (0.06304458401213482 * 2050) - 119.06594196773983
ax.plot([1880, 2050], [0, y2], color='red')
ynew = (0.1664272733318682 * 2050) - 325.7934668059649
ax.plot([2000, 2050], [7.062992, ynew], color='green')
plt.show()
```
